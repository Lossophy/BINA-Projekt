```python
# Datei einlesen
df = pd.read_csv('/content/BINA-Projekt/Data/ogd115_gest_bilanz.csv')

# 1. Daten vorbereiten
df_import = df[df["Rubrik"] == "Import"].copy()
df_inland = df[df["Rubrik"] == "Inlandproduktion"].copy()

df_import_sum = df_import.groupby("Jahr")["TJ"].sum().reset_index().rename(columns={"TJ": "Import"})
df_inland_sum = df_inland.groupby("Jahr")["TJ"].sum().reset_index().rename(columns={"TJ": "Inlandproduktion"})

# 2. Zusammenführen
df_combined = df_import_sum.merge(df_inland_sum, on="Jahr")
df_combined["Total"] = df_combined["Import"] + df_combined["Inlandproduktion"]

# 3. Prozentanteile berechnen
df_combined["Import (%)"] = df_combined["Import"] / df_combined["Total"] * 100
df_combined["Inland (%)"] = df_combined["Inlandproduktion"] / df_combined["Total"] * 100

# 4. Long-Format für Area Chart
df_plot = df_combined.melt(id_vars="Jahr", value_vars=["Import (%)", "Inland (%)"],
                           var_name="Quelle", value_name="Anteil (%)")

# 5. Plotten
fig = px.area(
    df_plot,
    x="Jahr",
    y="Anteil (%)",
    color="Quelle",
    title="Importabhängigkeit der Schweiz beim Energieverbrauch (1980–2023)",
    labels={"Quelle": "Energieherkunft"},
    groupnorm="percent"
)

fig.update_layout(template="plotly_white", yaxis_range=[0, 100], legend=dict(traceorder="reversed"))
fig.show()
```

#### Ursachen

Im Folgenden werden die identifizierten Trendbrüche und deren Ursachen dargelegt. Die Analyse der Gesamtenergiestatistik offenbart signifikante Trendbrüche im Energieimportverhalten der Schweiz, die sich über spezifische Zeiträume hinweg identifizieren lassen. Die vorliegenden Veränderungen stehen in einem engen Zusammenhang mit technologischen Entwicklungen, energiepolitischen Massnahmen sowie geopolitischen Rahmenbedingungen.

Ein erster signifikanter Trend manifestiert sich ab dem Jahr 2005: Seither ist ein kontinuierlicher Rückgang des Verbrauchs von Erdölprodukten zu beobachten. Diese Entwicklung ist primär auf Fortschritte in der Energieeffizienz, insbesondere im Automobil- und Heizsektor, sowie auf politische Massnahmen zur Dekarbonisierung zurückzuführen. Darüber hinaus beeinflusst die zunehmende Verbreitung der Elektromobilität die Nachfrage nach fossilen Brennstoffen.

Ein weiterer Wendepunkt lässt sich seit dem Jahr 2010 feststellen, als der Verbrauch von Kohleprodukten signifikant rückläufig war. Dieser Rückgang ist sowohl Ausdruck energie- und klimapolitischer Zielsetzungen als auch Resultat einer systematischen Substitution kohlebasierter Prozesse, insbesondere in der Industrie. Die geringe gesellschaftliche und politische Akzeptanz von Kohle sowie ihre vergleichsweise schlechte Umweltbilanz haben ihre Rolle im Energiemix marginalisiert.

Ein besonders aktueller und einschneidender Bruch ereignete sich ab dem Jahr 2022, als der Verbrauch von Gas deutlich zurückging. Die Ursachen hierfür sind in der geopolitischen Instabilität Europas zu verorten, konkret im Krieg in der Ukraine und der damit verbundenen Gasversorgungskrise. Diese Ereignisse haben die Verwundbarkeit der Schweiz im Bereich fossiler Energieimporte verdeutlicht und zu einer kurzfristigen Verhaltensanpassung auf politischer wie wirtschaftlicher Ebene geführt.

Schliesslich sind über den gesamten Zeitraum hinweg starke Schwankungen beim Elektrizitätsimport feststellbar. Diese Entwicklung lässt sich nicht auf einen einzelnen Trendbruch zurückführen, sondern resultiert aus strukturellen Rahmenbedingungen. Die saisonale Produktion von Wasserkraft, in Kombination mit der sogenannten Winterstromlücke, resultiert in einer verstärkten Importabhängigkeit der Schweiz während der Wintermonate. Diese Schwankungen sind Ausdruck eines Spannungsfeldes zwischen inländischer Produktionskapazität, wetterabhängiger Erzeugung und struktureller Importnotwendigkeit.

Die identifizierten Brüche können als Ausdruck eines sich wandelnden Energiesystems interpretiert werden, das zunehmend von Nachhaltigkeitszielen, technologischen Transformationen und äusseren Schocks geprägt ist.
```python
# 1. Rubriken und fossile Energieträger definieren
rubriken = [
    'Endverbrauch - Haushalte',
    'Endverbrauch - Dienstleistungen',
    'Endverbrauch - Industrie',
    'Endverbrauch - Statistische Differenz inkl. Landwirtschaft',
    'Endverbrauch - Verkehr'
]
fossil = ['Erdölprodukte', 'Gas', 'Kohle', 'Rohöl']

# 2. Filtern nach relevanten Rubriken und fossilen Energieträgern
df_fossil_rubriken = df[
    (df['Rubrik'].isin(rubriken)) &
    (df['Energietraeger'].isin(fossil))
].copy()

# 3. Gruppieren: Gesamtverbrauch fossiler Energieträger je Jahr & Rubrik
df_summe = df_fossil_rubriken.groupby(['Jahr', 'Rubrik'], as_index=False)['TJ'].sum()

# 3b. Rubriken umbenennen für bessere Lesbarkeit
rubrik_mapping = {
    'Endverbrauch - Haushalte': 'Haushalte',
    'Endverbrauch - Dienstleistungen': 'Dienstleistungen',
    'Endverbrauch - Industrie': 'Industrie',
    'Endverbrauch - Statistische Differenz inkl. Landwirtschaft': 'Differenz + Landwirtschaft',
    'Endverbrauch - Verkehr': 'Verkehr'
}
df_summe['Rubrik'] = df_summe['Rubrik'].replace(rubrik_mapping)

# 4. Nur Jahre mit Verbrauch > 0 behalten
df_summe = df_summe[df_summe['TJ'] > 0]

# 5. Sortieren für Darstellung
df_summe = df_summe.sort_values(by=['Jahr', 'TJ'], ascending=[True, False])

# 6. Farben definieren (optional)
color_map = {
    'Haushalte': '#1f77b4',
    'Dienstleistungen': '#ff7f0e',
    'Industrie': '#2ca02c',
    'Statistische Differenz + Landwirtschaft': '#d62728',
    'Verkehr': '#9467bd'
}

# 7. Bar Race erstellen
fig = px.bar(
    df_summe,
    x="TJ",
    y="Rubrik",
    animation_frame="Jahr",
    color="Rubrik",
    color_discrete_map=color_map,
    hover_name="Rubrik",
    range_x=[0, 305000],  # fixe X-Achsen-Grenze
    title='Gesamtverbrauch fossiler Energieträger nach Rubrik (TJ)',
    orientation='h'
)

# 8. Layout-Anpassungen
fig.update_layout(
    xaxis_title="Verbrauch in TJ",
    yaxis_title="Rubrik",
    legend_title="Rubrik",
    plot_bgcolor='white',
    hovermode='closest',
    showlegend=False
)

fig.update_xaxes(showgrid=False)
fig.update_yaxes(showgrid=False, categoryorder='total ascending')

# 9. Animation verlangsamen
fig.layout.updatemenus[0].buttons[0].args[1]['frame']['duration'] = 2000
fig.layout.updatemenus[0].buttons[0].args[1]['transition']['duration'] = 1000

# 10. Anzeigen
fig.show()
```

Anschließend erfolgte eine detaillierte Analyse der Entwicklung der absoluten Mengen, der relativen Anteile an der Summe beider Kategorien sowie des direkten Verhältnisses der inländischen erneuerbaren Produktion zu den fossilen Importen. Die Visualisierung dieser Entwicklungen erfolgte mittels Linien- und Flächendiagrammen, um Trends und Veränderungen über den betrachteten Zeitraum transparent darzustellen.

```python
# Datei einlesen
df = pd.read_csv('/content/BINA-Projekt/Data/ogd115_gest_bilanz.csv')

# Daten überprüfen
print(df.head())
```

```python
# Grundlegende Datenaufbereitung
df['TJ'] = pd.to_numeric(df['TJ'], errors='coerce').fillna(0)
df['Jahr'] = df['Jahr'].astype(int)

# --- Definitionen für F3 ---
erneuerbare_inland_basis_f3 = ["Holzenergie", "Wasserkraft", "Uebrige erneuerbare Energien"]
muell_energietraeger_f3 = "Müll und Industrieabfälle" # Einzelner String für leichtere Prüfung
# erneuerbare_inland_mit_muell_f3 wird jetzt in prepare_f3_data dynamischer gehandhabt
fossile_import_traeger_f3 = ["Erdölprodukte", "Gas", "Kohle", "Rohöl"]

# Farbcodierung für F3
farben_erneuerbar_inland_f3 = {
    "Holzenergie": "#1f77b4", "Wasserkraft": "#aec7e8", "Uebrige erneuerbare Energien": "#72bcd4",
    "Müll und Industrieabfälle (50% erneuerbar)": "#2ca02c" # Angepasster Name für Legende
}
farben_fossil_import_f3 = {
    "Erdölprodukte": "#ff7f0e", "Gas": "#ffbb78", "Kohle": "#d62728", "Rohöl": "#ff9896"
}

# color_map_f3_szenario1_detail bleibt wie es ist
color_map_f3_szenario1_detail = {
    **{k: v for k, v in farben_erneuerbar_inland_f3.items() if k != "Müll und Industrieabfälle (50% erneuerbar)"},
    **farben_fossil_import_f3
}

# color_map_f3_szenario2_detail wird angepasst, um den neuen Namen für Müll zu verwenden
color_map_f3_szenario2_detail = {**farben_erneuerbar_inland_f3, **farben_fossil_import_f3}


# --- Hilfsfunktion zur Datenaggregation (angepasst für 50% Müll) ---
def prepare_f3_data(df_input, erneuerbare_liste_basis, fossile_liste, muell_anteil_prozent=0, szenario_label=""):
    df_erneuerbare_inland = df_input[
        (df_input['Rubrik'] == "Inlandproduktion") &
        (df_input['Energietraeger'].isin(erneuerbare_liste_basis))
    ].copy() # .copy() um SettingWithCopyWarning zu vermeiden

    # Behandlung von Müll und Industrieabfälle
    df_muell_inland = pd.DataFrame() # Initialisieren für den Fall, dass muell_anteil_prozent = 0
    if muell_anteil_prozent > 0:
        df_muell_temp = df_input[
            (df_input['Rubrik'] == "Inlandproduktion") &
            (df_input['Energietraeger'] == muell_energietraeger_f3)
        ].copy()
        df_muell_temp['TJ'] = df_muell_temp['TJ'] * (muell_anteil_prozent / 100.0)
        df_muell_temp['Energietraeger'] = f"{muell_energietraeger_f3} ({muell_anteil_prozent}% erneuerbar)" # Für Legende
        df_muell_inland = df_muell_temp

    # Kombinieren der Basis-Erneuerbaren mit dem Müll-Anteil
    df_ern_inland_detail_final = pd.concat([df_erneuerbare_inland, df_muell_inland])

    # Detail-Daten für Plotting vorbereiten
    df_ern_inland_detail_grouped = df_ern_inland_detail_final.groupby(['Jahr', 'Energietraeger'])['TJ'].sum().reset_index()
    df_ern_inland_detail_grouped['Hauptkategorie'] = f'Inländ. Erneuerbare {szenario_label}'.strip()


    ern_inland_summe = df_ern_inland_detail_final.groupby('Jahr')['TJ'].sum().reset_index()
    ern_inland_summe = ern_inland_summe.rename(columns={'TJ': 'Erneuerbare_Inland_TJ'})

    df_foss_import_detail = df_input[
        (df_input['Rubrik'] == "Import") &
        (df_input['Energietraeger'].isin(fossile_liste))
    ].groupby(['Jahr', 'Energietraeger'])['TJ'].sum().reset_index()
    df_foss_import_detail['Hauptkategorie'] = 'Importe fossiler Energieträger'

    foss_import_summe = df_foss_import_detail.groupby('Jahr')['TJ'].sum().reset_index()
    foss_import_summe = foss_import_summe.rename(columns={'TJ': 'Fossile_Import_TJ'})

    df_detail_kombiniert_final = pd.concat([df_ern_inland_detail_grouped, df_foss_import_detail])

    df_vergleich = pd.merge(ern_inland_summe, foss_import_summe, on='Jahr', how='outer').fillna(0)
    df_vergleich = df_vergleich.sort_values(by='Jahr').reset_index(drop=True)

    df_vergleich['Verhaeltnis_ErnIn_vs_FossImp'] = df_vergleich['Erneuerbare_Inland_TJ'] / df_vergleich['Fossile_Import_TJ'].replace(0, np.nan)
    df_vergleich['Verhaeltnis_ErnIn_vs_FossImp'] = df_vergleich['Verhaeltnis_ErnIn_vs_FossImp'].fillna(0)

    summe_energien = df_vergleich['Erneuerbare_Inland_TJ'] + df_vergleich['Fossile_Import_TJ']
    label_suffix = f" {szenario_label}".rstrip()
    df_vergleich[f'Anteil_Erneuerbare_Inland_Prozent{label_suffix}'] = (df_vergleich['Erneuerbare_Inland_TJ'] / summe_energien.replace(0, np.nan) * 100).fillna(0)
    df_vergleich[f'Anteil_Fossile_Import_Prozent{label_suffix}'] = (df_vergleich['Fossile_Import_TJ'] / summe_energien.replace(0, np.nan) * 100).fillna(0)

    return df_detail_kombiniert_final, df_vergleich

# --- Daten für Szenarien vorbereiten ---
# Szenario 1: Basis Erneuerbare (ohne Müll)
df_detail_s1_f3, df_vergleich_s1_f3 = prepare_f3_data(df, erneuerbare_inland_basis_f3, fossile_import_traeger_f3, muell_anteil_prozent=0, szenario_label="(Basis)")

# Szenario 2: Erneuerbare INKLUSIVE 50% von Müll und Industrieabfälle
df_detail_s2_f3, df_vergleich_s2_f3 = prepare_f3_data(df, erneuerbare_inland_basis_f3, fossile_import_traeger_f3, muell_anteil_prozent=50, szenario_label="(inkl. 50% Müll)")
```

```python
def plot_f3_s2_absolut_aggregiert_tab():
    if df_vergleich_s2_f3.empty: return go.Figure().update_layout(title_text="Daten für Szenario 2 nicht geladen") # Geändert auf df_vergleich_s2_f3
    fig = go.Figure()
    fig.add_trace(go.Scatter(x=df_vergleich_s2_f3['Jahr'], y=df_vergleich_s2_f3['Erneuerbare_Inland_TJ'],
                             mode='lines+markers', name='Inländ. Erneuerbare (inkl. 50% Müll)', line=dict(color=farben_erneuerbar_inland_f3["Müll und Industrieabfälle (50% erneuerbar)"]))) # Legendenname und Farbe angepasst
    fig.add_trace(go.Scatter(x=df_vergleich_s2_f3['Jahr'], y=df_vergleich_s2_f3['Fossile_Import_TJ'],
                             mode='lines+markers', name='Importe fossiler Energieträger', line=dict(color=farben_fossil_import_f3["Erdölprodukte"])))

    fig.update_layout(
        title_text="Abs. Mengen (aggregiert): Erneuerbare (inkl. 50% Müll) vs. Fossile Importe", # Titel angepasst
        title_x=0.5,
        xaxis_title="Jahr",
        yaxis_title="Energiemenge (TJ)",
        hovermode="x unified",
        legend_title_text='Kategorie',
        legend=dict(traceorder="reversed"))
    return fig

beschreibung_s2_absolut_aggregiert = """
**Grafik: Absolute Mengen (aggregiert): Inl. Erneuerbare vs. Fossile Importe (Szen. Müll)**

Diese Grafik zeigt die absolute Entwicklung der Energiemengen (in Terajoule, TJ) über die Jahre.
Verglichen werden die Summe der inländisch produzierten erneuerbaren Energien mit der Summe der importierten fossilen Energieträger.
"""
```

```python
def plot_f3_s2_absolut_detail_tab():
    if df_detail_s2_f3.empty: return go.Figure().update_layout(title_text="Daten für Szenario 2 nicht geladen") # Geändert auf df_detail_s2_f3
    # Erstelle eine benutzerdefinierte Sortierreihenfolge für die Legende und Stapelung
    kategorie_reihenfolge_detail_s2 = [et for et in erneuerbare_inland_basis_f3]
    kategorie_reihenfolge_detail_s2.append("Müll und Industrieabfälle (50% erneuerbar)") # Angepasster Name
    kategorie_reihenfolge_detail_s2.extend(fossile_import_traeger_f3)

    fig = px.area(df_detail_s2_f3, x="Jahr", y="TJ", color="Energietraeger", line_group="Hauptkategorie", # line_group beibehalten oder anpassen, falls nötig
                  color_discrete_map=color_map_f3_szenario2_detail,
                  title="Abs. Mengen (detailliert): Erneuerbare (inkl. 50% Müll) vs. Fossile Importe", # Titel angepasst
                  labels={"TJ": "Energiemenge (TJ)", "Energietraeger": "Energieträger"},
                  category_orders={"Energietraeger": kategorie_reihenfolge_detail_s2})

    fig.update_layout(
        title_x=0.5,
        legend_title_text='Energieträger',
        hovermode="x unified",
        legend=dict(traceorder="reversed"))
    return fig

beschreibung_s2_absolut_detail = """
**Grafik: Absolute Mengen (detailliert): Inl. Erneuerbare vs. Fossile Importe (Szen. Müll)**

Dieses Flächendiagramm stellt die absoluten Energiemengen (in Terajoule, TJ) der einzelnen Energieträger über die Jahre dar.
Die Flächen sind gestapelt, wobei die obere Gruppe die inländischen erneuerbaren Energien und die untere Gruppe die importierten fossilen Energieträger detailliert aufschlüsselt.
"""
```

```python
def plot_f3_s2_anteil_100prozent_aggregiert_tab():
    if df_vergleich_s2_f3.empty: return go.Figure().update_layout(title_text="Daten für Szenario 2 nicht geladen")
    df_anteile_melted = df_vergleich_s2_f3.melt(id_vars=['Jahr'],
                                              value_vars=['Anteil_Erneuerbare_Inland_Prozent (inkl. 50% Müll)', 'Anteil_Fossile_Import_Prozent (inkl. 50% Müll)'], # Angepasste Spaltennamen
                                              var_name='Kategorie_Anteil', value_name='Prozentualer_Anteil')
    df_anteile_melted['Kategorie_Anteil'] = df_anteile_melted['Kategorie_Anteil'].replace({
        'Anteil_Erneuerbare_Inland_Prozent (inkl. 50% Müll)': 'Anteil Inländ. Erneuerbare (inkl. 50% Müll)', # Angepasst
        'Anteil_Fossile_Import_Prozent (inkl. 50% Müll)': 'Anteil Fossile Importe'
    })
    fig = px.area(df_anteile_melted, x="Jahr", y="Prozentualer_Anteil", color="Kategorie_Anteil",
                  title="Anteilige Entwicklung (aggregiert, 100% Ansicht, inkl. 50% Müll)", # Titel angepasst
                  labels={"Prozentualer_Anteil": "Prozentualer Anteil (%)"},
                  color_discrete_map={'Anteil Inländ. Erneuerbare (inkl. 50% Müll)': farben_erneuerbar_inland_f3["Müll und Industrieabfälle (50% erneuerbar)"], 'Anteil Fossile Importe': '#ff7f0e'}, # Farbe angepasst
                  groupnorm='percent')

    fig.update_layout(
        title_x=0.5,
        legend_title_text='Energiekategorie',
        hovermode="x unified",
        legend=dict(traceorder="reversed"))
    fig.update_yaxes(ticksuffix="%")
    return fig

beschreibung_s2_anteil_100_aggregiert = """
**Grafik: Anteilige Entwicklung (aggregiert, 100% Ansicht, Szen. Müll)**

Dieses gestapelte Flächendiagramm zeigt die prozentuale Verteilung zwischen der Summe der inländischen erneuerbaren Energien und der Summe der importierten fossilen Energieträger über die Jahre.
Die Y-Achse summiert sich für jedes Jahr auf 100%, wodurch die relativen Anteile der beiden Hauptkategorien direkt vergleichbar werden.
"""
```

```python
def plot_f3_s2_verhaeltnis_tab():
    if df_vergleich_s2_f3.empty: return go.Figure().update_layout(title_text="Daten für Szenario 2 nicht geladen") # Geändert auf df_vergleich_s2_f3
    fig = px.line(df_vergleich_s2_f3, x="Jahr", y="Verhaeltnis_ErnIn_vs_FossImp",
                  title="Verhältnis Inl. Erneuerbare (inkl. 50% Müll) zu Fossilen Importen", # Titel angepasst
                  labels={"Verhaeltnis_ErnIn_vs_FossImp": "Verhältnis"}, markers=True,
                  color_discrete_sequence=[farben_erneuerbar_inland_f3["Müll und Industrieabfälle (50% erneuerbar)"]]) # Farbe hinzugefügt
    fig.add_hline(y=1, line_dash="dash", line_color="grey", annotation_text="Gleichstand")

    fig.update_layout(
        title_x=0.5,
        hovermode="x unified",
        legend=dict(traceorder="reversed"))
    return fig

beschreibung_s2_verhaeltnis = """
**Grafik: Verhältnis Inl. Erneuerbare zu Fossilen Importen (Szen. Müll)**

Dieses Liniendiagramm stellt das Verhältnis der Energiemenge aus inländischer erneuerbarer Produktion zu der Menge der importierten fossilen Energieträger dar.
Ein Wert über 1 (gestrichelte Linie) bedeutet, dass mehr erneuerbare Energie im Inland produziert als fossile Energie importiert wurde; ein Wert unter 1 das Gegenteil.
"""
```

```python
def plot_f3_s2_anteil_summe_tab():
    if df_vergleich_s2_f3.empty:
        return go.Figure().update_layout(title_text="Daten für Szenario 2 nicht geladen")

    fig = px.line(df_vergleich_s2_f3,
                  x='Jahr',
                  y='Anteil_Erneuerbare_Inland_Prozent (inkl. 50% Müll)', # Angepasster Spaltenname
                  title='Anteil Inl. Erneuerbare (inkl. 50% Müll) an (Inl. Erneuerbare (inkl. 50% Müll) + Fossile Importe)', # Titel angepasst
                  labels={'Anteil_Erneuerbare_Inland_Prozent (inkl. 50% Müll)': 'Anteil (%)'}, # Label angepasst
                  markers=True,
                  line_shape='linear',
                  color_discrete_sequence=[farben_erneuerbar_inland_f3["Müll und Industrieabfälle (50% erneuerbar)"]]) # Passende Farbe

    fig.update_layout(
        title_x=0.5,
        xaxis_title="Jahr",
        yaxis_title="Anteil (%)",
        hovermode="x unified",
        yaxis_ticksuffix="%",
        legend=dict(traceorder="reversed")
    )
    return fig

beschreibung_s2_anteil_summe = """
**Grafik: Anteil Inl. Erneuerbare an Summe (Inl. Ern. + Foss. Imp., Szen. Müll)**

Dieses Liniendiagramm zeigt den prozentualen Anteil, den die inländisch produzierten erneuerbaren Energien an der Gesamtmenge aus (inländischen Erneuerbaren + fossilen Importen) über die Jahre ausmachen.
Die Y-Achse gibt diesen Anteil in Prozent an.
"""
```

```python
def plot_f3_s2_anteil_100prozent_detail_tab():
    if df_detail_s2_f3.empty:
        return go.Figure().update_layout(title_text="Daten für Szenario 2 (inkl. 50% Müll) nicht geladen") # Titel angepasst

    df_detail_s2_f3_sorted = df_detail_s2_f3.copy()
    # Erstelle eine benutzerdefinierte Sortierreihenfolge für die Legende und Stapelung
    kategorie_reihenfolge_detail_s2 = [et for et in erneuerbare_inland_basis_f3]
    kategorie_reihenfolge_detail_s2.append("Müll und Industrieabfälle (50% erneuerbar)") # Angepasster Name
    kategorie_reihenfolge_detail_s2.extend(fossile_import_traeger_f3)

    df_detail_s2_f3_sorted['Energietraeger'] = pd.Categorical(
        df_detail_s2_f3_sorted['Energietraeger'],
        categories=kategorie_reihenfolge_detail_s2,
        ordered=True
    )
    df_detail_s2_f3_sorted = df_detail_s2_f3_sorted.sort_values(by=['Jahr', 'Energietraeger'])

    fig = px.area(df_detail_s2_f3_sorted,
                  x="Jahr",
                  y="TJ",
                  color="Energietraeger",
                  line_group="Energietraeger",
                  groupnorm='percent',
                  title="Anteilige Entwicklung (detailliert, 100% Ansicht): Erneuerbare (inkl. 50% Müll) vs. Fossile Importe", # Titel angepasst
                  labels={"TJ": "Prozentualer Anteil (%)", "Energietraeger": "Energieträger"},
                  color_discrete_map=color_map_f3_szenario2_detail, # Verwendet die aktualisierte color_map
                  category_orders={"Energietraeger": kategorie_reihenfolge_detail_s2}
                 )

    fig.update_layout(
        title_x=0.5,
        legend_title_text='Energieträger',
        hovermode="x unified",
        yaxis_ticksuffix="%",
        legend=dict(traceorder="reversed")
    )
    return fig

beschreibung_s2_anteil_100_detail = """
**Grafik: Anteilige Entwicklung (detailliert, 100% Ansicht, Szen. Müll)**

Dieses gestapelte Flächendiagramm visualisiert die prozentuale Zusammensetzung der betrachteten Energiemixe (Inländische Erneuerbare und Importe fossiler Energieträger) über die Jahre, aufgeschlüsselt nach einzelnen Energieträgern.
Die Y-Achse summiert sich für jedes Jahr auf 100% und zeigt so die relativen Beiträge der einzelnen Energieträger innerhalb der beiden Hauptkategorien.
"""
```

Die Analyse zeigt, dass der Anteil erneuerbarer Energien aus inländischer Produktion im Verhältnis zu den Importen fossiler Energieträger zwar gestiegen ist, die Abhängigkeit von fossilen Importen jedoch weiterhin erheblich ist. Im Kontext der Energiestrategie 2050, die eine Reduktion der energiebedingten Umweltbelastung und eine Stärkung der Versorgungssicherheit durch inländische erneuerbare Energien anstrebt, lassen sich aus der fiktiven Perspektive von Mitarbeitern des Bundesamtes für Energie (BFE) folgende Handlungsempfehlungen ableiten, um die Transformation zu beschleunigen:

- Verstärkte Förderung des Ausbaus diversifizierter erneuerbarer Energien:
Neben der Wasserkraft, die bereits einen hohen Anteil liefert, sollte der Fokus weiterhin stark auf dem Ausbau von Photovoltaik, Windenergie, Biomasse (inkl. der energetischen Nutzung des biogenen Anteils von Abfällen) und Geothermie liegen. Die Daten deuten auf ein Wachstum bei den "Uebrigen erneuerbaren Energien" hin, dieses Potenzial gilt es konsequent weiter zu erschließen.

- Konkretisierung BFE-Perspektive:
Entwicklung und Anpassung von Förderinstrumenten, um Investitionsanreize zu optimieren und administrative Hürden für den Ausbau zu senken. Stärkung der Forschung und Entwicklung in zukunftsträchtigen erneuerbaren Technologien.

- Beschleunigung der Substitution fossiler Energieträger:
Insbesondere im Wärme- und Verkehrssektor, wo fossile Energien noch stark dominieren **(nicht direkt Teil von F3, aber relevanter Kontext -> Bezug zu F2 nennen)**, müssen Substitutionspfade durch erneuerbare Alternativen und Effizienzmaßnahmen aggressiver verfolgt werden.

- Konkretisierung BFE-Perspektive:
Ausarbeitung und Kommunikation von Roadmaps für die Dekarbonisierung verschiedener Sektoren. Förderung von Infrastruktur für erneuerbare Mobilität und Wärme (z.B. Ladeinfrastruktur, Wärmenetze).

- Verbesserung der Rahmenbedingungen für Speichertechnologien und Netzausbau: Eine höhere Volatilität durch den Ausbau von Solar- und Windenergie erfordert Investitionen in Speicherlösungen und einen intelligenten Netzausbau, um die Versorgungssicherheit zu gewährleisten und die Integration erneuerbarer Energien zu optimieren.

- Konkretisierung BFE-Perspektive:
Schaffung regulatorischer Anreize für Speicherinvestitionen. Koordination des Netzausbaus unter Berücksichtigung des dezentralen Einspeisepotenzials.

- Kontinuierliches Monitoring und transparente Kommunikation:
Die Entwicklung des Verhältnisses von inländischen Erneuerbaren zu fossilen Importen sollte weiterhin engmaschig überwacht und die Fortschritte (oder Defizite) im Hinblick auf die Ziele der Energiestrategie 2050 transparent kommuniziert werden, um den politischen und gesellschaftlichen Handlungsdruck aufrechtzuerhalten.

- Konkretisierung BFE-Perspektive:
Regelmäßige Veröffentlichung von Fortschrittsberichten und verständlichen Visualisierungen (wie die für diese Case Study entwickelten) für Politik, Wirtschaft und Öffentlichkeit.

#### Fazit

Die Analyse der Schweizerischen Gesamtenergiestatistik von 1980 bis 2023 zeigt eine positive, wenngleich langsame Entwicklung des Anteils inländischer erneuerbarer Energien (inklusive 50% des energetisch genutzten Mülls) gegenüber den Importen fossiler Energieträger. Das Verhältnis hat sich zugunsten der Erneuerbaren verbessert, und ihr prozentualer Anteil an der Summe beider Kategorien ist gestiegen. Dennoch bleibt die Schweiz substanziell von fossilen Energieimporten abhängig. Um die ambitionierten Ziele der Energiestrategie 2050 – insbesondere die Reduktion der energiebedingten Umweltbelastung – zu erreichen, bedarf es weiterhin konsequenter und beschleunigter Maßnahmen zur Förderung erneuerbarer Energien und zur Reduktion des Verbrauchs fossiler Energieträger. Die hier vorgeschlagenen Handlungsempfehlungen zielen darauf ab, diesen Transformationsprozess aus der Perspektive des BFE zu unterstützen.

#### Analyse und Beantwortung der Frage

Die Frage nach der Entwicklung des Anteils erneuerbarer Energien aus inländischer Produktion im
Vergleich zu den Importen fossiler Energieträger ist von zentraler Bedeutung für das Verständnis
der Schweizer Energiewende und der Fortschritte im Rahmen der Energiestrategie 2050. Diese
Strategie zielt maßgeblich darauf ab, die energiebedingte Umweltbelastung zu reduzieren und die
Abhängigkeit von oft volatilen und klimaschädlichen fossilen Energieimporten zu verringern.


#### Methodischer Ansatz

Für die Beantwortung dieser Frage wurde die Schweizerische Gesamtenergiestatistik für den
Zeitraum von 1980 bis 2023 analysiert.
Als "erneuerbare Energien aus inländischer Produktion" wurden die Energieträger Holzenergie,
Wasserkraft sowie die Sammelkategorie "Übrige erneuerbare Energien" (die typischerweise
Solarthermie, Photovoltaik, Umweltwärme, Biogas etc. umfasst) aus der Rubrik
"Inlandproduktion" aggregiert. Der Posten "Müll und Industrieabfälle" wurde hier bewusst nicht
pauschal den erneuerbaren Energien zugerechnet, da nur dessen biogener Anteil erneuerbar ist
und der Datensatz keine detailliertere Aufschlüsselung bietet. Eine separate Betrachtung dieses
Postens kann jedoch interessante Zusatzinformationen liefern.
Die "Importe fossiler Energieträger" setzen sich aus den importierten Mengen von
Erdölprodukten, Rohöl (als Vorprodukt für fossile Energieträger), Gas und Kohle zusammen
(Rubrik "Import").
Die Analyse fokussiert auf die zeitliche Entwicklung der absoluten Mengen dieser beiden
Aggregate sowie, zur direkten Beantwortung der Fragestellung, auf die Entwicklung ihres
Verhältnisses und des prozentualen Anteils der erneuerbaren Inlandsproduktion an der Summe
beider Aggregate.


#### Entwicklung der Energiemengen und ihres Verhältnisses

Betrachtet man die absoluten Energiemengen über den Zeitraum von 1980 bis 2023, so wird
deutlich, dass die Importe fossiler Energieträger über die gesamte Periode hinweg mengenmäßig
dominierten. Innerhalb der fossilen Importe stellten Erdölprodukte und Rohöl traditionell die
größten Posten dar, wobei Gasimporte im Zeitverlauf an Bedeutung gewannen, während die
Kohleimporte tendenziell zurückgingen.
Die inländische Produktion erneuerbarer Energien, traditionell stark von der Wasserkraft geprägt,
zeigte ebenfalls eine Zunahme. Insbesondere Holzenergie und die "Übrigen erneuerbaren
Energien" verzeichneten vor allem ab den 2000er Jahren einen signifikanten relativen Zuwachs,
wenngleich von einer absolut niedrigeren Basis ausgehend.
Die entscheidende Entwicklung zur Beantwortung der Fragestellung zeigt sich jedoch im
Verhältnis und den Anteilen. Obwohl die fossilen Importe in absoluten Zahlen stets höher lagen,
hat sich der Anteil der erneuerbaren Energien aus inländischer Produktion im direkten Vergleich zu
den fossilen Importen tendenziell verbessert. Visualisierungen, die diesen relativen Vergleich
darstellen (z.B. als Verhältnislinie oder als 100%-Stapeldiagramm der beiden Aggregate),
verdeutlichen eine langsame, aber stetige Verschiebung zugunsten der heimischen Erneuerbaren.
Dies bedeutet, dass die Schweiz, relativ gesehen, ihren Bedarf (im Rahmen der hier betrachteten
Aggregate) zunehmend durch inländische erneuerbare Quellen im Vergleich zu fossilen Importen
deckt.
Ein Verhältniswert von "Erneuerbar Inland / Fossil Import" über 1, der bedeuten würde, dass mehr
erneuerbare Energie im Inland produziert als fossile Energie importiert wird, wurde im
Betrachtungszeitraum nicht erreicht. Der Trend zeigt jedoch eine Annäherung an diesen Punkt
bzw. eine Verringerung der relativen Dominanz der fossilen Importe.


#### Einflussfaktoren und energiepolitische Kontextualisierung

Diese Entwicklung ist vor dem Hintergrund verschiedener energiepolitischer und geopolitischer
Ereignisse zu sehen:

##### Frühe Phase (1980er - 1990er):
Eine hohe Grundabhängigkeit von importierten fossilen
Energieträgern, insbesondere Öl, prägte diese Zeit. Energieeﬃzienz und erneuerbare
Energien standen noch nicht im primären Fokus.

##### Jahrtausendwende und frühe 2000er:
Ein wachsendes Bewusstsein für Klimawandel und
die Einführung erster Förderinstrumente (z.B. Einspeisevergütungen) begünstigten langsam
den Ausbau neuer erneuerbarer Energien. Ereignisse wie das Tschernobyl-Nachbeben und
die Diskussionen um die Kernenergie verstärkten das Interesse an Alternativen.

##### Energiestrategie 2050 (ab ca. 2011 beschlossen, schrittweise umgesetzt):
Dieses
strategische Rahmenwerk hat klare Ziele zur Reduktion des Verbrauchs fossiler Energien
und zum Ausbau erneuerbarer Energien formuliert. Die Daten ab diesem Zeitraum sollten
idealerweise eine beschleunigte Verschiebung der Anteile zeigen. Die Zunahme der
"Übrigen erneuerbaren Energien" ist hier ein deutlicher Indikator.

##### Geopolitische Ereignisse:
Der Ukraine-Krieg und die damit verbundene Gaskrise ab 2022
haben die Verletzlichkeit durch Energieimporte drastisch vor Augen geführt und die
Dringlichkeit einer Stärkung der inländischen (erneuerbaren) Energieproduktion sowie der
Diversifizierung nochmals erhöht. Die aktuellsten Datenpunkte bis 2023 könnten erste
Reaktionen hierauf widerspiegeln, beispielsweise in Form einer erhöhten Sensibilität für den
Gasverbrauch oder verstärkten Bemühungen im Bereich der erneuerbaren Energien.

##### Technologischer Fortschritt und Kostenreduktion:
Insbesondere bei Photovoltaik und
teilweise auch Windkraft haben gesunkene Kosten den Ausbau attraktiver gemacht.


#### Bewertung im Kontext der Energiestrategie 2050

Die beobachtete graduelle Zunahme des Anteils erneuerbarer Energien aus inländischer
Produktion im Verhältnis zu den fossilen Importen ist grundsätzlich positiv und steht im Einklang
mit den Zielen der Energiestrategie 2050. Sie zeigt, dass politische Maßnahmen und
gesellschaftliche Veränderungen eine Wirkung entfalten.
Allerdings verdeutlicht die weiterhin hohe absolute Menge an fossilen Importen auch die
Dimension der Herausforderung. Um die ambitionierten Ziele der Energiestrategie – insbesondere
eine drastische Reduktion der CO2-Emissionen und der Importabhängigkeit – zu erreichen, ist
eine deutliche Beschleunigung der Transformation notwendig. Der bisherige Trend ist ein Schritt in
die richtige Richtung, aber die Geschwindigkeit des Wandels muss zunehmen. Die Analyse des
Szenarios, bei dem "Müll und Industrieabfälle" (zumindest deren biogener Anteil) zu den
erneuerbaren Energien gezählt werden, würde den Anteil der "erneuerbaren" Seite zwar erhöhen,
die grundlegende Herausforderung der Reduktion fossiler Importe bleibt jedoch bestehen.


#### Fazit

Über den Zeitraum von 1980 bis 2023 hat sich der Anteil erneuerbarer Energien aus inländischer
Produktion gegenüber den Importen fossiler Energieträger in der Schweiz tendenziell positiv
entwickelt. Während die Schweiz weiterhin maßgeblich von fossilen Energieimporten abhängig ist,
ist der relative Beitrag der heimischen erneuerbaren Energien gestiegen. Diese Entwicklung
spiegelt die Bemühungen wider, die im Rahmen der Energiestrategie 2050 angestoßen wurden,
und wird durch externe Faktoren wie technologischen Fortschritt und geopolitische
Notwendigkeiten beeinflusst. Die vollständige Erreichung der strategischen Ziele erfordert jedoch
weiterhin erhebliche Anstrengungen und eine beschleunigte Transformation des Energiesystems,
insbesondere im Hinblick auf die Reduktion des Gesamtverbrauchs fossiler Energien und den
konsequenten Ausbau aller Potenziale der erneuerbaren Energien im Inland.

Wir sind der Meinung, dass wir mit den gezeigten Fragestellungen und Visualisierungen praxisnah demonstrieren können, wie aus den Daten der Schweizerischen Gesamtenergiestatistik resp. dem Kernstück Energiebilanz nützliche und adressatengerechte Einblicke in historische und aktuelle Daten zur Energiewirtschaft möglich und wichtig sind. In den Analysen und Handlungsempfehlungen haben wir herausgearbeitet, welche Aussagen sich damit treffen lassen und welche (bereits beschlossenen) Massnahmen gestärkt werden können.

- Erneuerbare Energieträger weiter ausbauen und fördern

- Modernere (effizientere) Technologien fördern und Rahmenbedingungen verbessern
- Lösungen zur besseren Speicherung weiter erforschen und unterstützen

Energieträger

Elektrizität:  

Sekundärenergieträger, umgewandelt aus Primärenergieträgern wie Erdöl (Rohöl), Erdgas, Kohle, Torf, Natururan bzw. Kernenergie, Holz und andere Biomasse, Wind-, Gezeiten-, Wellen-, Meeresströmungs- und Wasserkraft, Sonnenstrahlung, Erd- und Umgebungswärme sowie Müll und Industrieabfälle. Manchmal wird mehr als ein Umwandlungsprozess durchlaufen, z.B. Stromerzeugung aus bereits raffinierten Produkten wie Benzin/Diesel.

Erdölprodukte:

Sekundärenergieträger wie Heizöle (leicht bis schwer), Benzin, Diesel(öl), Flugtreibstoffe, Petrolkoks und übrige energetische Erdölprodukte (Flüssiggase (z.B. Propan/Butan), Leuchtpetrol, White Spirit («Testbenzin», dient als Löse- und Verdünnungsmittel in Produkten wie Klebstoffen und Anstrichen wie z. B. Holzschutzmitteln, Lacken sowie Dispersions- und Ölfarben.), und VGO (Vakuumgasöl, Zwischenprodukt des Raffinationsprozesses von Erdöl)).

Fernwärme:  

Sekundärenergieträger, meint Wärme, die an Dritte verkauft wird und für deren Haupttransport und Verteilung öffentlicher Boden beansprucht wird. Kann auch bei Kernkraftwerken, Müllverbrennungsanlagen, etc. produziert werden.

Gas:  

Umfasst Inlandproduktion (vorübergehend zwischen 1985 und 1994 in Finsterwald) und Import von Erdgas, Erzeugung aus Kohle und Erdölprodukten, Biogaseinspeisung ins Erdgasnetz. Der Endverbrauch berechnet sich nach Abzügen der teilweisen Umwandlung in Elektrizität und Fernwärme sowie den Eigenverbrauch der Gaswerke und Netzverluste.

Holzenergie:

Wird über 20 verschiedene Verbrennungsanlagentypen erhoben. Diese können grob in Einzelraumheizungen, Gebäudeheizungen oder automatische Feuerungen eingeteilt werden.

Kernbrennstoffe:

Primärenergieträger wie Natururan, zur Elektrizitätserzeugung in Kernkraftwerken genutzt.

Kohle:

Gemeint sind Kohlenarten wie Steinkohle (sowie Steinkohlenbriketts oder -koks) oder Braunkohle. Die Holzkohle ist hier nicht gemeint.

Müll und Industrieabfälle:

Primärenergieträger, gemeint ist die energetische Nutzung von Abwärme, die bei der Verbrennung von Müll (Kehricht in Kehrichtverbrennungsanlagen) und Industrieabfällen (z.B. Klärschlämme oder Abfälle aus der Papierindustrie).  

Rohöl:

Primärenergieträger, gemeint ist Erdöl. In den Raffinerien werden daraus Sekundärenergieträger wie etwa Erdölprodukte.

Übrige erneuerbare Energien:

Zusammenfassung von Biogas (aus Mist und Gülle der Landwirtschaft, vergorenen Haushaltabfällen, Abfalldeponien, Kläranlagen und Industrieabwässern gewonnen), biogenen Treibstoffen (Biodiesel, Bioethanol und pflanzliche/tierische Öle/Altöle), Sonne (Elektrizität mittels Photovoltaik und Warmwasser mittels Sonnenkollektoren), Wind (Leistung und Umweltwärme (mittels Wärmepumpen aus Luft, Oberflächenwasser, Grundwasser oder Erdwärme)

Wasserkraft:

Primärenergieträger, gemeint ist die in den Wasserkraftwerken erzeugten Elektrizitätsmenge z.B. in Lauf-, Speicher- oder Pumpspeicherwerken.



Sektoren  

Verkehr:  

Umfasst den gesamten privaten und öffentlichen Verkehr (Strasse, Schiene, Wasser und Luft) ohne weitere Untergliederung nach Sektoren (Industrie, Dienstleistungen, etc.)

Haushalte:  

In der Regel privater Verbrauch von Energie für Raumwärme, Warmwasserbereitung, Beleuchtung, etc.

Industrie:

Umfasst auch das verarbeitende Gewerbe. Folgt seit 1999 einer repräsentativen Stichprobe von tausenden Arbeitsstätten anhand der Wirtschaftssystematik NOGA, welche für einige Energieträger entsprechend hochgerechnet wird. Zuvor durch einen Energiekonsumentenverband erhoben.  

Dienstleitungen:

Siehe Angaben Industrie zur Erhebung nach der Wirtschaftssystematik NOGA.  

Differenz + Landwirtschaft:  

Statistische Differenz meint die Differenz zwischen der gemessenen Abgabe an Endverbraucher und den Einzelmeldungen je Endkundenkategorie.
